```python
import numpy as geek
from datetime import datetime, timedelta

Day  = geek.arange(813)
Month =geek.arange(813)
Year = geek.arange(813)
Date_sum = geek.arange(813)

for i in range(0,813):
    date = dataset.iloc[i,0]
    split_date = date.split('/')
    new_date = [eval(i) for i in split_date]
    month= new_date[0]
    day= new_date[1]
    year= new_date[2]
    Day[i] = day
    Month[i] = month
    Year[i] = year
    date_str = datetime(year=year, month=month, day=day)
    Date_sum[i] = (date_str.year*10000 +date_str.month * 100 +date_str.day * 1)
    
```

```python
dataset['Day'] = Day
dataset['Month'] = Month
dataset['Year'] = Year
dataset['Date_sum'] = Date_sum
```
